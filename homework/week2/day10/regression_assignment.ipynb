```python
## with test train split and cross validation 

from sklearn import metrics

def rmse(y_true, y_pred):
    return np.sqrt(metrics.mean_squared_error(y_true, y_pred)
                  )

def adjusted_r2(y_true, y_pred, n, k):
    r2 = metrics.r2_score(y_true, y_pred)
    return 1 - (1-r2)*(n-1) / (n - k - 1)
```

```python
from sklearn.linear_model import LinearRegression

model_ols = LinearRegression()

model_ols.fit(X= independent_variables, y= target_variable)

model_ols_preds = model_ols.predict(independent_variables)
```

```python
RESULTS = {}

def evaluate_model(y_true, y_pred, n, k):
    return {
        "rmse": rmse(y_true, y_pred),
        "mae": metrics.mean_absolute_error(y_true, y_pred),
        "adjusted_r2": adjusted_r2(y_true, y_pred, n, k)
           }
```

```python
N = independent_variables.shape[0]

RESULTS["ols"] = evaluate_model(
    target_variable,
    model_ols_preds,
    N,
    len(model_ols.coef_)
)

RESULTS
```
