### Multiple Group Counts

```python
def pivot_count(df, rows, columns, calc_field):
    df_pivot = df.pivot_table(values=calc_field, 
                              index=rows, 
                              columns=columns, 
                              aggfunc=np.size
                             ).dropna(axis=0, how='all')
    return df_pivot
```

```python
price_by_size = pivot_count(airbnb, "accommodation_types", "room_type", "price")
price_by_size
```

```python
def heatmap_count_types(df, col1, col2):
    pivot_table = pivot_count(df,col1, col2, "price")
    sns.heatmap(pivot_table, annot=True, fmt='g')
    plt.ylabel(col1)
    plt.xlabel(col2)
    plt.show()
```

```python
interact(heatmap_count_types, col1=airbnb.columns,
                                 col2=airbnb.columns,
                                 df=fixed(airbnb));
```

### Categorical Variable Means 

```python
def means_by_category(col_group, col_calc):
    airbnb_no_outliers = airbnb[airbnb[col_calc]<airbnb[col_calc].quantile(0.99)]
    airbnb_no_outliers.groupby(col_group)[col_calc].mean().plot.barh()   
    plt.ylabel(col_group)
    plt.xlabel('Mean values for variable {}'.format(col_calc))
    plt.show()
```

```python
numerical_variables = airbnb.select_dtypes(np.number).columns
categorical_variables = airbnb.select_dtypes(['object', 'category']).columns
columns_type = [col for col in airbnb.columns if col.endswith("_type")]
```

```python
interact(means_by_category, col_group=categorical_variables, col_calc=numerical_variables);
```

### Means by Multiple Groups

```python
def pivot_mean(rows, columns, calc_field):
    df_pivot = airbnb.pivot_table(values=calc_field, 
                              index=rows, 
                              columns=columns, 
                              aggfunc=np.mean
                             ).dropna(axis=0, how='all')
    return df_pivot
```

```python
pivot_mean("room_type", "accommodation_types", "price")
```

# Analysis Conclusion

### Description

The data set contains airbnb listings throughout different neighborhoods. Information on size, price, and number of bedrooms is included. 

The entites associated to the dataset are:
- price
- price-location
- price-bedrooms/accomodates
- reviews
- satisfaction

## Data Dictionary 

The variables that exist on the dataset are:

- room_id(int). Identifies unique listing.
- host_id(int). Identifies unique hosts.
- room_type(categorical). Description of listing.
- neighborhood(categorical).
- reviews(int). Number of reviews recieved
- overall_satisfaction(int). Satisfaction rating
- accommodates(int). Number of people listing accommodates.
- bedrooms(int). Number of bedrooms.
- price(int).

## QA

- There are no duplicate rows. 
- A majority of the listings are for the 'entire home'.
- A majority of the listings are from a few neighborhoods.
- A majority of the listings accommodate around 1-8 people, while a smaller portion accommodates over 10 people. 
- Most are 1-3 bedrooms. 
- Reviews counts the numbers of reviews given, a large portion has 0, this variable needs to be adjusted. 
- The most apparent outliers are in the 'price' variable

## Variable Distribution

- No numerical variables follow a normal distribution 
- A majority of the listings are in a few neighborhoods, with Santa Maria Maior having over 20%.
- About 35% of the listings are considered very small based on how many the accommodate. 
- The satisfaction type variable is almost evenly distributed across all categories. 
- The price type variable is also almost evenly distributed. 

## Comparisons

- Entire homes are assocated with a higher price compared to options like shared rooms. 
- Increased number of bedrooms/ accomodation size is consistent with higher prices.
- Parque de Nacoes has higher average prices than any other neighborhood.
